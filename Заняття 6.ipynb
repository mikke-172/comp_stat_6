```python
import numpy as np
from numpy import random as npr
from matplotlib import pyplot as plt
```

```python
n = 10000
M = 1000
```

```python
x = npr.normal(0, 1, n)
plt.hist(x, bins=20)
```

```python
def sigma(x, b):
    x = np.array(x)
    return 1 / (1 + np.exp(-b[0] - b[1] * x))

b = [-0.5, 3]
sigma([-1, 1, 2, 3], b)
```

```python
t = np.linspace(-3, 3, 100)

plt.plot(t, sigma(t, b))
```

```python
plt.scatter(x, sigma(x, b), alpha=0.01)
```

```python
sx = sigma(x, b)
```

```python
sx.mean(), (sx ** 2).mean(), (sx ** 3).mean()
```

```python
x = npr.normal(0, 1, (n, M))
```

```python
plt.hist(sigma(x, b).mean(axis=0))
```

```python
results = sigma(x, b).mean(axis=0)
```

```python
b, results.mean()
```

```python
(results ** 2).mean(), (results ** 3).mean()
```
